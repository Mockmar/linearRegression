```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
df
```

```python
km = np.array(df['km'])
km
```

```python
price = np.array(df['price'])
price
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='km', y='price', data=df)
plt.title('Scatter plot of x and y')
plt.show()
```

```python
def normalize_km(km):
    return (km - km.mean()) / km.std()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=normalize_km(km), y=price)
plt.title('Scatter plot of x and y')
plt.show()
a = 0
b = 0

L = 0.0001
epochs = 1000
n= float(len(km))

km = normalize_km(km)

for i in range(epochs):
    y_pred = a*km + b
    D_a = (-2/n) * sum(km * (price - y_pred))
    D_b = (-2/n) * sum(price - y_pred)
    a = a - L * D_a
    b = b - L * D_b
    plt.figure(figsize=(10, 6))
    plt.plot(km, price, 'ro')
    plt.plot([min(km), max(km)], [min(y_pred), max(y_pred)], color='blue')
    plt.title('Regression line')
    plt.show()

print(a, b)
```
